```python
string = input("Enter a string: ")
freq = {}

for ch in string:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```

```python
string = input("Enter a string: ")

frequency = {}

for char in string:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Character Frequencies:")
for key, value in frequency.items():
    print(key, ":", value)
```

```python
s = input("Enter a string: ")

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
