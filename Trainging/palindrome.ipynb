```python
s = input("Enter a string: ")

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
def check(n):

    for i in range(n):
        for j in range(n):

            if i != j and i*i + j*j == n:
                return True

    return False


n = int(input())

print(check(n))
```
