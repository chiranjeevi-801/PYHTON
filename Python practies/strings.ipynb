### Count Vowels in a String

```python
a="chiranjeevi"
count=0

for ch in a:
    if ch in "aeiouAEIOU":
        count+=1
print("vowels:",count)
```

### Palindrome

```python
a="madam"

if a == a[::-1]:
    print("Palindrom")
else:
    print("not a Plaindrom")
```

```python
a=(input("enter the word:"))

if a == a[::-1]:
    print("Palindrom")
else:
    print("not a Plaindrom")
```

```python
a=(input("enter the words"))

a=a.lower()

if a == a[::-1]:
    print("Palindrom")
else:
    print("not a Plaindrom")
```

### Check Anagram

```python
s1="listen"
s2="silent"

if sorted(s1)==sorted(s2):
    print("Anagram")
else:
    print("Not Anagram")
```
