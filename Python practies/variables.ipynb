### reverse a number 


```python
num = 1234
reverse= 0 

while num > 0:
    digit = num %10 
    reverse = reverse *10 + digit 
    num = num //10 

print("Reversed:",reverse)
```

```python
num = 1234
reverse = int(str(num)[::-1])
print(reverse)
```

```python
num = 1234
reverse = 0
num_str = str(num) # We use the string length to know how many loops to run

for _ in range(len(num_str)):
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reversed:", reverse)
```

### Armstrong Number

```python
num=153
temp =num 
total = 0 

while temp >0:
    digit=temp%10
    total+=digit**3
    temp//=10
if total == num:
    print("Armstrong Number")
else:
    print("Not a Armstrong")
```

### palindrome 

```python
num = 121
s=str(num)
if s=="".join(reversed(s)):
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
num=int(input("enter the number"))
s=str(num)
if s=="".join(reversed(s)):
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
num= 121 
temp=num
reverse=0

while temp>0:
    digit =temp%10
    reverse=reverse*10+digit
    temp//=10

if reverse==num:
    print("Palindrome")
else:
    print("Not a Palindrome")
```
