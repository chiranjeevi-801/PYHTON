### 3. Loop Statement

Repeat work.

### For loop

```python
for i in range(5):
    print(i)
```

```python
fruits = ["Apple", "Banana", "Cherry"]

for fruit in fruits:
    print("I need to buy a(n):", fruit)
```

```python
numbers=[10,20,30,40]

total=0

for n in numbers:
    total+=n  # This adds the current number to our running total

print(total)
```

```python
word= "PYTHON"

for letter in word:
    print("Current letter:",letter)
```

```python
# Print only even numbers from 0 to 10
for i in range(0, 11, 2):
    print(i)
```

### While loop

```python
i=0
while i<5:
    print(i)
    i+=1
```

```python
timer=5

while timer>0:
    print("seconds remaining:",timer)
    timer-=1
print("Blast of!  🚀")
```

```python
password = ""

while password != "Alvas123":
    password = input("Enter the password: ")
    
    if password != "Alvas123":
        print("Access Denied. Try again.")


print("Access Granted!")

    
```

```python
current_weight = int(input("enter the current_weight"))
limit = 50

while current_weight < limit:
    add_weight = 15
    current_weight += add_weight
    print("Current load:", current_weight, "kg")

print("The truck is full!")
```

```python
current_weight = 0
limit = 50

while current_weight < limit:
    add_weight = 15
    current_weight += add_weight
    print("Current load:", current_weight, "kg")

print("The truck is full!")
```

### 4. Function Statement

Reusable code block.

```python
def greet():
    print("Hello")
greet()
```

```python
def greet_person(name):
    print("Hello, " + name + "!")

greet_person("Alice")
greet_person("Bob")
```

```python
def add_numbers(a,b):
    return a+b

result=add_numbers(10,10)
print("The sum is:",result)
```

```python
def fire_rocket(count):
    print("Rocket #" + str(count) + " launched! 🚀")

for i in range(1, 4):
    fire_rocket(i)
```
