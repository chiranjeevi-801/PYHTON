### 3) ATM PIN Verification and Withdrawal Program

```python
pin = 1234
balance = 5000
count = 0

while count < 4:
    amount = int(input("Enter PIN: "))

    if amount == pin:
        wdr = int(input("Enter amount to withdraw: "))

        if wdr <= balance:
            print("Successful")
            print("Balance:", balance - wdr)
        else:
            print("Insufficient Balance")
        break

    else:
        count = count + 1
        print("Wrong PIN")

if count == 4:
    print("Too many wrong PIN attempts.")
    print("Please wait for 24 hours or contact your bank.")
```

### 5) WAP to read a name from user, seperate vowels and consonant in the  name 

```python
name=input("enter a name:")
vowels="aeiouAEIOU"
vowels_list=[]
consonant_list=[]
for ch in name :
    if ch.isalpha():
        if ch in vowels:
            vowels_list.append(ch)
        else:
            consonant_list.append(ch)
print("vowels:",vowels_list)
print("consonant:",consonant_list)
```

### 8) squaring a list (numbers)


```python
numbers = [1, 2, 3, 4, 5]

square = []

for i in numbers:
    square.append(i ** 2)

print(square)


```

```python
numbers = [1, 2, 3, 4, 5]

square = [i ** 2 for i in numbers]

print(square)

```

```python
numbers = [1, 2, 3, 4, 5]

square = list(map(lambda x: x ** 2, numbers))

print(square)
```

### 9) extract even list 

```python
numbers = [1, 2, 3, 4, 5]

even = [i for i in numbers if i % 2 == 0]

print(even)
```

### Using filter  function extract even list 

```python
numbers = [1, 2, 3, 4, 5]

even = list(filter(lambda x: x % 2 == 0, numbers))

print(even)
```

 ### 10) Recurcive Function

```python
def fact(n):
    if n==0:
        return 1
    else:
        return n*fact(n-1)
print(fact(5))
    
```

### 11) Rational numbers 

```python
def fact(n):
    if n==0:
        return 1
    else:
        return n+fact(n-1)
print(fact(5))
    
```
