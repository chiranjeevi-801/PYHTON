### Find largest number 

```python
arr = [10, 20, 30, 40, 50]

largest = arr[0]

for i in range(len(arr)):
    if arr[i] > largest:
        largest = arr[i]

print("Largest element is:", largest)
```

### find second largest number 

```python

arr = [10, 20, 30, 40, 50]

largest = second = 0

for i in arr:
    if i > largest:
        second = largest
        largest = i
    elif i > second:
        second = i

print("Second largest element:", second)
```

```python
arr = [10, 20, 30, 40, 50]

n = len(arr)

l = arr[0]

for i in range(1, n):
    if arr[i] > l:
        l = arr[i]

s = arr[0]

for i in range(1, n):
    if arr[i] > s and arr[i] < l:
        s = arr[i]

print("Second largest element:", s)
```

### Write a Python program to read a name from the user, print the name, and change every alternate character to the next alphabetical character.

```python
name = input("Enter your name: ")

result = ""

for i, ch in enumerate(name):
    if i % 2 == 1:        
        result += chr(ord(ch) + 1)
    else:
        result += ch

print("Original Name :", name)
print("Modified Name :", result)
```

### Write a Python program to split an array into two equal halves. Check whether both halves are equal. If they are equal, check whether the numbers are even or odd.

```python
arr = [1, 5, 1, 3, 5, -1]

mid = len(arr) // 2

fh = sum(arr[:mid])
sh = sum(arr[mid:])

if fh == sh:
    print("Both halves are equal")

    if fh % 2 == 0:
        print("Sum is Even")
    else:
        print("Sum is Odd")
else:
    print("Both halves are not equal")
```

### Tranpose matrix 

```python
r = int(input("Enter number of rows: "))
c = int(input("Enter number of columns: "))

matrix = []

print("Enter the matrix elements:")
for i in range(r):
    row = list(map(int, input().split()))
    matrix.append(row)

print("\nOriginal Matrix:")
for i in range(r):
    for j in range(c):
        print(matrix[i][j], end=" ")
    print()

print("\nTranspose Matrix:")
for i in range(c):
    for j in range(r):
        print(matrix[j][i], end=" ")
    print()
```

### Adding of two matrix 

```python
r = int(input("Enter number of rows: "))
c = int(input("Enter number of columns: "))
matrix1 = []
matrix2 = []
print("Enter elements of First Matrix:")
for i in range(r):
    row = list(map(int, input().split()))
    matrix1.append(row)
print("Enter elements of Second Matrix:")
for i in range(r):
    row = list(map(int, input().split()))
    matrix2.append(row)
print("\nFirst Matrix:")
for i in range(r):
    for j in range(c):
        print(matrix1[i][j], end=" ")
    print()
print("\nSecond Matrix:")
for i in range(r):
    for j in range(c):
        print(matrix2[i][j], end=" ")
    print()
print("\nSum of the Matrices:")
for i in range(r):
    for j in range(c):
        print(matrix1[i][j] + matrix2[i][j], end=" ")
    print()
```

### Subtraction of Matrices

```python
r = int(input("Enter rows: "))
c = int(input("Enter columns: "))

a = []
b = []

print("Enter First Matrix:")
for i in range(r):
    a.append(list(map(int, input().split())))

print("Enter Second Matrix:")
for i in range(r):
    b.append(list(map(int, input().split())))

print("Subtraction of Matrices:")

for i in range(r):
    for j in range(c):
        print(a[i][j] - b[i][j], end=" ")
    print()
```
