### Find largest number 

```python
arr = [10, 20, 30, 40, 50]

largest = arr[0]

for i in range(len(arr)):
    if arr[i] > largest:
        largest = arr[i]

print("Largest element is:", largest)
```

### find second largest number 

```python

arr = [10, 20, 30, 40, 50]

largest = second = 0

for i in arr:
    if i > largest:
        second = largest
        largest = i
    elif i > second:
        second = i

print("Second largest element:", second)
```

```python
arr = [10, 20, 30, 40, 50]

n = len(arr)

l = arr[0]

for i in range(1, n):
    if arr[i] > l:
        l = arr[i]

s = arr[0]

for i in range(1, n):
    if arr[i] > s and arr[i] < l:
        s = arr[i]

print("Second largest element:", s)
```
