### Find largest number 

```python
arr = [10, 20, 30, 40, 50]

largest = arr[0]

for i in range(len(arr)):
    if arr[i] > largest:
        largest = arr[i]

print("Largest element is:", largest)
```

### find second largest number 

```python

arr = [10, 20, 30, 40, 50]

largest = second = 0

for i in arr:
    if i > largest:
        second = largest
        largest = i
    elif i > second:
        second = i

print("Second largest element:", second)
```

### Write a Python program to read a number from the user and print the next greater number from the array

```python
arr = [2, 3, 4, 5, 7, 9, 11, 12, 15]

num = int(input("Enter a number: "))

for i in arr:
    if i > num:
        print("Next greater number is:", i)
        break
```

### Write a Python program to read a name from the user, print the name, and change every alternate character to the next alphabetical character.

```python
name = input("Enter your name: ")

result = ""

for i, ch in enumerate(name):
    if i % 2 == 1:        
        result += chr(ord(ch) + 1)
    else:
        result += ch

print("Original Name :", name)
print("Modified Name :", result)
```

### Write a Python program to split an array into two equal halves. Check whether both halves are equal. If they are equal, check whether the numbers are even or odd.

```python
arr = [1, 5, 1, 3, 5, -1]

mid = len(arr) // 2

fh = sum(arr[:mid])
sh = sum(arr[mid:])

if fh == sh:
    print("Both halves are equal")

    if fh % 2 == 0:
        print("Sum is Even")
    else:
        print("Sum is Odd")
else:
    print("Both halves are not equal")
```
