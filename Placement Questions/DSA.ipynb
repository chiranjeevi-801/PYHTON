### Tranpose matrix 

```python
r = int(input("Enter number of rows: "))
c = int(input("Enter number of columns: "))

matrix = []

print("Enter the matrix elements:")
for i in range(r):
    row = list(map(int, input().split()))
    matrix.append(row)

print("\nOriginal Matrix:")
for i in range(r):
    for j in range(c):
        print(matrix[i][j], end=" ")
    print()

print("\nTranspose Matrix:")
for i in range(c):
    for j in range(r):
        print(matrix[j][i], end=" ")
    print()
```

# Kadane's Algorithm

```python
arr = [1,3,2,4,5]

sum = 0
max = arr[0]

for i in arr:
    sum += i

    if sum > max:
        max = sum

    if sum < 0:
        sum = 0

print("Largest Sum:", sum)
```

####  Write a Python program to count all subarrays of length k where k is even and the subarray is even-dominated.

```python
arr = [2, 4, 1, 6, 3, 8]

k = 4
count = 0

for i in range(len(arr) - k + 1):

    even = 0
    odd = 0

    for j in range(i, i + k):
        if arr[j] % 2 == 0:
            even += 1
        else:
            odd += 1

    if even > odd:
        count += 1

print("Count:", count)
```

```python
arr = [2,3,5,4,3,6,7,1,2]
k = 4

count = 0
even = 0


for i in range(k):
    if arr[i] % 2 == 0:
        even += 1

if even > k // 2:
    count += 1


for i in range(k, len(arr)):
    if arr[i-k] % 2 == 0:
        even -= 1

    if arr[i] % 2 == 0:
        even += 1

    if even > k // 2:
        count += 1

print("Count:", count)
```

### A Stack is a linear data structure that follows the LIFO (Last In, First Out) principle.

(or)

### A stack is a linear data structure in which the last inserted element is the first element to be removed (LIFO).

Stack

Top

│

30  ← Removed first

20

10

│

Bottom

Types of Stack:---

### 1.Static Stack:

    Fixed size.
    
    Cannot grow after creation.


### 2.Dynamic Stack:

   Size can increase or decrease.

   Python lists implement a dynamic stack.

### Stack Operations:



Stack: It is the fundamental of linear data structure

It is used to store collection of ordered elements.


push()--->inserting an element into the stack

pop()--->deleting an element from the stack

peek()/top()---> view the top element

display()--->used to display the stack element

isEmpty()---->Check whether the stack is empty

Size()---->Find the number of elements

if it is full then print"Overflow"

while deleting an element, check whether the stack is empty or not.

if it is empty then print"Underflow"
