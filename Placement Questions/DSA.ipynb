### Find largest number 

```python
arr = [10, 20, 30, 40, 50]

largest = arr[0]

for i in range(len(arr)):
    if arr[i] > largest:
        largest = arr[i]

print("Largest element is:", largest)
```

### find second largest number 

```python

arr = [10, 20, 30, 40, 50]

largest = second = 0

for i in arr:
    if i > largest:
        second = largest
        largest = i
    elif i > second:
        second = i

print("Second largest element:", second)
```

```python
arr = [10, 20, 30, 40, 50]

n = len(arr)

l = arr[0]

for i in range(1, n):
    if arr[i] > l:
        l = arr[i]

s = arr[0]

for i in range(1, n):
    if arr[i] > s and arr[i] < l:
        s = arr[i]

print("Second largest element:", s)
```

### Write a Python program to read a number from the user and print the next greater number from the array

```python
arr = [2, 3, 4, 5, 7, 9, 11, 12, 15]

num = int(input("Enter a number: "))

for i in arr:
    if i > num:
        print("Next greater number is:", i)
        break
```

### Write a Python program to read a name from the user, print the name, and change every alternate character to the next alphabetical character.

```python
name = input("Enter your name: ")

result = ""

for i, ch in enumerate(name):
    if i % 2 == 1:        
        result += chr(ord(ch) + 1)
    else:
        result += ch

print("Original Name :", name)
print("Modified Name :", result)
```
