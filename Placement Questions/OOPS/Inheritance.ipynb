## What is Inheritance?

### Inheritance means one class acquires the properties and methods of another class.

Simple Definition :--

### Inheritance allows one class (child class) to use the properties and methods of another class (parent class).

It helps reuse code instead of writing the same code again.

# Types of Inheritance
   





##  1. Single Inheritance :--
      
  #### One parent and one child.

          Animal
            ↑
           Dog


## 2. Multiple Inheritance :--

### One child inherits from multiple parents.


   Father      Mother

      \     /
       Child

## 3. Multilevel Inheritance



Grandparent → Parent → Child

 Animal

       ↑

  Mammal

       ↑

  Dog





 # (or)






 Grand father 

     ↑

 father

     ↑

  son

## 4. Hierarchical Inheritance



### One parent has multiple children.




       Animal
      /      \
    Dog      Cat



## (OR)


       Father 
      /      \
 child-1     Child-2

## 5. Hybrid Inheritance



### Hybrid inheritance is a combination of two or more types of inheritance (for example, multiple + multilevel).

        Animal
       /      \
    Mammal   Bird
        \     /
         Bat

# 1) Single Inheritance 

```python
class Father:
    def prop1(self):
        print("father's property")
class Son:
    def prop2(self):
        print("Son's property")
        
f = Father()
f.prop1()
s= Son()
s.prop2()
```

```python
class Father:
    def prop1(self):
        print("father's property")
class Son(Father):
    def prop2(self):
        print("Son's property")
        
# f = Father()
# f.prop1()
s= Son()
s.prop2()
s.prop1()
```
