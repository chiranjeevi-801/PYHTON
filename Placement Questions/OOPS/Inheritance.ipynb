### Inheritance means one class acquires the properties and methods of another class.

Simple Definition :--

### Inheritance allows one class (child class) to use the properties and methods of another class (parent class).

It helps reuse code instead of writing the same code again.

# 2. Multilevel Inheritance

Definition: A child inherits from a parent, and another child inherits from that child.

```python
class GFather:
    def prop1(self):
        print("Grand father's property")
class Father(GFather):
    def prop2(self):
        print("Father's property")
class Son(Father):
    def prop3(self):
        print("son's property")
        
s= Son()
s.prop3()
s.prop2()
s.prop1()
```

```python
class GFather:
    def prop1(self):
        print("Grandfather's property")

class Father(GFather):
    def prop2(self):
        print("Father's property")

class Son(Father):
    def prop3(self):
        print("Son's property")

s = Son()

s.prop1()
s.prop2()
s.prop3()
```

# 3. Multiple Inheritance

Definition: One child class inherits from two or more parent classes.

```python
class Father:
    def prop1(self):
        print("Father's property")

class Mother:
    def prop2(self):
        print("Mother's property")

class Son(Father, Mother):
    def prop3(self):
        print("Son's property")

s = Son()

s.prop1()
s.prop2()
s.prop3()
```

# 4. Hierarchical Inheritance
Definition: One parent class has multiple child classes.

```python
class Father:
    def prop1(self):
        print("Father's property")

class Son(Father):
    def prop2(self):
        print("Son's property")

class Daughter(Father):
    def prop3(self):
        print("Daughter's property")

s = Son()
d = Daughter()

s.prop1()
s.prop2()

d.prop1()
d.prop3()
```

# 5. Hybrid Inheritance

Definition: Hybrid inheritance is a combination of two or more inheritance types.

Example: Multilevel + Multiple

```python
class GFather:
    def prop1(self):
        print("Grandfather's property")

class Father(GFather):
    def prop2(self):
        print("Father's property")

class Mother:
    def prop3(self):
        print("Mother's property")

class Son(Father, Mother):
    def prop4(self):
        print("Son's property")

s = Son()

s.prop1()
s.prop2()
s.prop3()
s.prop4()
```

Write a Python program to implement abstraction using the abc module. Create an abstract class Vehicle and demonstrate how different vehicles (Car, Bike, and Scooty) implement the abstract method start() differently.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass


class Car(Vehicle):

    def start(self):
        print("Car starts with a Key.")


class Bike(Vehicle):

    def start(self):
        print("Bike starts with a Kick.")


class Scooty(Vehicle):

    def start(self):
        print("Scooty starts with Self Start.")



c = Car()
b = Bike()
s = Scooty()


c.start()
b.start()
s.start()
```

Create an abstract class Vehicle with the methods start() and tyres(). Derive the classes Car, Bike, and Scooty, implement the methods, create objects, and display the output.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def tyres(self):
        pass


class Car(Vehicle):

    def start(self):
        print("Car starts with a Key.")

    def tyres(self):
        print("Car has 4 tyres.")


class Bike(Vehicle):

    def start(self):
        print("Bike starts with a Kick.")

    def tyres(self):
        print("Bike has 2 tyres.")


class Scooty(Vehicle):

    def start(self):
        print("Scooty starts with Self Start.")

    def tyres(self):
        print("Scooty has 2 tyres.")


c = Car()
b = Bike()
s = Scooty()

print(" Car ")
c.start()
c.tyres()

print("\n Bike ")
b.start()
b.tyres()

print("\n Scooty ")
s.start()
s.tyres()
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def tyres(self):
        pass


class Car(Vehicle):
    def start(self):
        print("Car starts with Key")

    def tyres(self):
        print("number of 4 tyres")


class Bike(Vehicle):
    def start(self):
        print("Bike starts with Kick")

    def tyres(self):
        print("number of 2 tyres")


class Scooty(Vehicle):
    def start(self):
        print("Scooty starts with Self Start")

    def tyres(self):
        print("number 2 tyres")


for vehicle in [Car(), Bike(), Scooty()]:
    vehicle.start()
    vehicle.tyres()
    print()
```
