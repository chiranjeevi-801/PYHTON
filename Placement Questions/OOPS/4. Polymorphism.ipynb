## 4. Polymorphism




Polymorphism means one method can have different implementations in different classes.

## Example 1 : Different Classes, Same Method

```python
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

d = Dog()
c = Cat()

d.sound()
c.sound()
```

## Example 2: Vehicle

```python

class Car:
    def start(self):
        print("Car starts with Key")

class Bike:
    def start(self):
        print("Bike starts with Kick")

c = Car()
b = Bike()

c.start()
b.start()
```
