# What is a Class?

#### A class is a blueprint or template used to create objects.





Simple Definition

#### A class is a blueprint that defines the properties (variables) and behaviors (functions) of an object.

## Real-Life Example

Imagine a Car Factory.

The factory designs a car.

But until a car is manufactured, it's just a design.

That design is the Class.

## Syntax

### class Student:
### pass


Here,

### Student is a class.

# What is an Object?

### An object is a real instance of a class.



Simple Definition

### An object is a real-world entity created from a class.

### syntax

class Student:

    pass

s1 = Student()

Here,

Student → Class

s1 → Object

# example

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(self.name)
        print(self.age)

s1 = Student("Rahul", 21)

s1.display()
```

or

```python
class Car:

    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def show(self):
        print("Brand :", self.brand)
        print("Color :", self.color)

c1 = Car("BMW", "Black")

c1.show()
```
