## 2. Abstraction:


Abstraction means showing only the essential details and hiding the implementation.


## Animal


```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Dog Barks")

d = Dog()
d.sound()
```

## Example: Payment System


```python
from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self):
        pass

class UPI(Payment):
    def pay(self):
        print("Paid using UPI")

u = UPI()
u.pay()
```
