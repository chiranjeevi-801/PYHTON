# 1. What is OOPS? 
#### (OOPS stands for Object-Oriented Programming System.)


### It is a programming method where we create objects from classes to solve real-world problems.



Simple Definition

## OOPS is a programming concept that organizes code using classes and objects.

Car → Class

My Car → Object

### Why do we use OOPS?

Code becomes reusable

Easy to understand

Easy to maintain

Reduces code duplication

Models real-world objects

Class

Object

Inheritance

Note: Many people say "types of OOPS," but these are actually the main concepts (pillars) of OOP.

#### Interview Answer

## Q: What are the four pillars of OOP?

The four pillars of Object-Oriented Programming are:

Encapsulation – Wrapping data and methods into a class.

Abstraction – Hiding implementation details and showing only essential features.

Inheritance – Reusing properties and methods of another class.

Polymorphism – Allowing the same method to perform different actions in different classes.
