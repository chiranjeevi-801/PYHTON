## Encapsulation :


Encapsulation is the process of combining data (variables) and methods (functions) into a single unit (class) and protecting the data from direct access

```python
class Student:
    def __init__(self):
        self.__marks = 90

    def show(self):
        print(self.__marks)

s = Student()
s.show()
```

## Bank Account

```python
class Bank:
    def __init__(self):
        self.__balance = 5000

    def show(self):
        print("Balance:", self.__balance)

b = Bank()
b.show()
```
